# NumPy array Indexing/Slicing

```python
from si.io.csv_file import read_csv
import numpy as np
```

```python
# 1.1 read csv

file = "/home/sa_bruno/Documentos/GitHub/si/datasets/iris/iris.csv"
dataset = read_csv(file, features= True, label=True)
print(dataset.summary())
print()
print(dataset.shape())
dataset.X
```

```python
#1.2 Penultimate independent variable/ dimension

penultimate_feature = dataset.X[:,-2]
penultimate_feature.shape
```

```python
#1.3 Last 10 samples / mean of the last 10 samples for each independent variable/feature

last_10_samples = dataset.X[-10:]
print(last_10_samples)
last_10_samples.mean(axis=0)
```

```python
# 1.4  Select all samples from the dataset with values less than or equal to 6 for all independent variables/features. How many samples do you obtain?

filtered_samples = dataset.X[(dataset.X <= 6).all(axis=1)]
filtered_samples.shape[0]
```

```python
# 1.5  Select all samples with a class/label different from 'Iris-setosa'

filtered_not_setosa = dataset.X[(dataset.y != "Iris-setosa")]
filtered_not_setosa.shape[0]
```
